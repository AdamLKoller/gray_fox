```python
import pandas as pd
```

```python
def get_center(l):
    l = l.split('],[')[1:-1]
    l = [y.split(',') for y in l]
    l = [[float(y[0]), float(y[1])] for y in l]
    
    sum_x = sum([x[0] for x in l])
    mean_x = sum_x / len(l)
    
    sum_y = sum([y[1] for y in l])
    mean_y = sum_y / len(l)
    
    
    return [mean_x, mean_y]
```
