```python
bamID_to_sampleID = pd.read_csv('./data/meta/bamID_to_sampleID.csv',header=None)
replicates = bamID_to_sampleID.loc[bamID_to_sampleID[1].str.endswith('replicate')]
originals = bamID_to_sampleID.loc[bamID_to_sampleID[1].isin([x.split('replicate')[0] for x in list(replicates[1])])]
pd.concat([originals, replicates], axis=0)
```

```python
combos = []
for sample in originals[1]:
    combos.append((sample, sample+'replicate'))
```
